```python
#function to convert emojis into word 

def convert_emojis(text):
    for emot in UNICODE_EMO:
        text = re.sub(r'('+emot+')', "_".join(UNICODE_EMO[emot].replace(",","").split()))
        
        return text
```

```python
def review_text(text):
    
    #1. convert words to lower 
    text = text.lower()
    
    #2. remove weblinks 
    text =re.sub(r"https\S+","",str(text))
    
    #3. remove tags
    text =re.sub(r"@\S+","",str(text))
    
    #4. remove tags
    text = convert_emojis(text)
    
    #5. tokenize words 
    tokenizer = RegexpTokenizer('[a-z]\w+')
    words = tokenizer.tokenize(text)
    
    #6. lemmatize words 
    lem_words = [lemmatizer.lemmatize(i) for i in words]
    
    #7. convert stopwords to a set, as searching a set is faster than than a list
    stops = stopwords.words('english')
    meaningful_words = [w for w in lem_words if not w in stops]
    
    return (" ".join(meaningful_words))
```
